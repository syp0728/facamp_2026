## Rolling Stats per Team

Purpose: For each match, compute a team's historical performance stats using only data from before that match (no data leakage), then return those as new columns.

Iterates per team — for each team, it isolates and time-sorts that team's matches.

Two types of lookback, both with shift(1) to exclude the current game:

- Expanding (full cycle): gs.shift(1).expanding().mean() — average over all prior matches since Aug 2018. Columns: goals_avg, goals_suf_avg, rank_avg, points_avg
- Rolling (last 5 games): gs.shift(1).rolling(5, min_periods=1).mean() — average over the most recent 5 matches. Columns: goals_avg_l, goals_suf_avg_l, rank_avg_l, points_avg_l

### Generate

```python
def rolling_team_stats(
    df, team_col, goals_scored_col, goals_conceded_col,
    rank_col, points_col, window=5
):
    # compute per-team rolling stats
    # - full cycle
    # - last `n` games (window)

    rows = []
    teams = df[team_col].unique()

    for team in teams:

        mask = df[team_col] == team
        team_df = df[mask].sort_values('date').copy()

        gs = team_df[goals_scored_col]
        gc = team_df[goals_conceded_col]
        rk = team_df[rank_col]
        pts = team_df[points_col]

        # expanding (full cycle) — shift(1) so we never use the current game's data
        team_df['goals_avg'] = gs.shift(1).expanding().mean()
        team_df['goals_suf_avg'] = gc.shift(1).expanding().mean()
        team_df['rank_avg'] = rk.shift(1).expanding().mean()
        team_df['points_avg']  = pts.shift(1).expanding().mean()

        # rolling (last `n` games)
        team_df['goals_avg_lastn'] = gs.shift(1).rolling(window, min_periods=1).mean()
        team_df['goals_suf_avg_lastn'] = gc.shift(1).rolling(window, min_periods=1).mean()
        team_df['rank_avg_lastn'] = rk.shift(1).rolling(window, min_periods=1).mean()
        team_df['points_avg_lastn'] = pts.shift(1).rolling(window, min_periods=1).mean()

        rows.append(team_df)

    result = pd.concat(rows).sort_index()
    
    return result
```

```python
df_stats = rolling_team_stats(
    df_concat,
    team_col='team',
    goals_scored_col='goals_scored',
    goals_conceded_col='goals_conceded',
    rank_col='opp_rank',
    points_col='team_points'
)
```
