# Merge

```python
import pandas as pd
```

## Load

```python
df_stats = pd.read_parquet("./intermediate/stats.parquet")
```

```python
df_stats
```

```python
# index refers to the same match (home vs. away = away vs. home)
df_stats.groupby(level=0).size()
```

```python
df_stats.loc[1348]
```

```python
basic_cols = ["date", "team", "opponent", "goals_scored"]

stat_cols = [
    'goals_avg', 'goals_avg_lastn',
    'goals_suf_avg', 'goals_suf_avg_lastn',
    'rank_avg', 'rank_avg_lastn',
    'points_avg', 'points_avg_lastn'
]
```

```python
basic_cols + stat_cols
```

## Home Team

```python
df_stats[ df_stats.is_home == 1].loc[:, basic_cols + stat_cols]
```

```python
df_home = df_stats[ df_stats.is_home == 1].loc[:, basic_cols + stat_cols].copy()
```

## Away Team

```python
df_stats[ df_stats.is_home == 0].loc[:, basic_cols + stat_cols]
```

```python
df_away = df_stats[ df_stats.is_home == 0].loc[:, basic_cols + stat_cols].copy()
```

## Sanity Check

```python
df_home.loc[1348]
```

```python
df_away.loc[1348]
```

```python
df_home_away = pd.merge(
    df_home, df_away,
    left_on=["date", "team", "opponent"],
    right_on=["date", "opponent", "team"],
    suffixes=["_home", "_away"],
    how="inner"
)
```

```python
df_home_away
```

```python
# rename
df_home_away.rename(
    columns={
        "team_home": "home_team",
        "opponent_home": "away_team"
    },
    inplace=True
)
```

```python
# drop unnecessary
df_home_away.drop(
    columns=["team_away", "opponent_away"],
    inplace=True
)
```

```python
df_home_away
```

```python
df_home_away.to_parquet("./intermediate/features.parquet")
```
