# Feature Engineering (Part 2)

````python
import pandas as pd
import os
````

## Stats Difference

### Compute

````python
df_features = pd.read_parquet("./intermediate/features.parquet")
````

````python
df_features
````

````python
stat_cols = [
    'goals_avg', 'goals_avg_lastn',
    'goals_suf_avg', 'goals_suf_avg_lastn',
    'rank_avg', 'rank_avg_lastn',
    'points_avg', 'points_avg_lastn'
]
````

````python
df_features["goals_avg_home"]  - df_features["goals_avg_away"]
````

````python
for stat_col in stat_cols:
    col_home = f"{stat_col}_home"  # goals_avg_home
    col_away = f"{stat_col}_away"  # goals_avg_away
    col_diff = f"{stat_col}_diff"  # goals_avg_diff
    df_features[col_diff] = df_features[col_home]  - df_features[col_away]
````

````python
df_features
````

### Sanity Check

````python
df_features.loc[1224]
````

## Rank Difference

### Compute

````python
df_match_rank = pd.read_parquet(
    "./intermediate/match_rank.parquet",
    columns=[
        "date",
        "home_team", "away_team",
        "home_score", "away_score",
        "rank_home", "rank_away"
    ]
)
````

````python
df_match_rank
````

````python
df_match_rank["rank_diff"] = df_match_rank["rank_home"] - df_match_rank["rank_away"]
````

````python
df_match_rank
````

### Sanity Check

````python
df_match_rank.loc[1348]
````

````python
41 - 28
````

## Merge Stats Difference and Rank Difference

````python
df_final = pd.merge(
    df_features,
    df_match_rank,
    on=["date", "home_team", "away_team"],
    how="left"
)
````
