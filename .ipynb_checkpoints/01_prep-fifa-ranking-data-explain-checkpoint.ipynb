# Prep FIFA Ranking Data

```python
import pandas as pd
import os
```

```python
pd.__version__
```

## Load FIFA Ranking Data

`fifa_ranking-2022-10-06.csv` contains FIFA ranking data from 1992-12-31 to 2022-10-06.

```python
rank = pd.read_csv('../data/fifa_ranking-2022-10-06.csv')
```

```python
rank
```

## Convert to DateTime

```python
rank['rank_date'] = pd.to_datetime(rank['rank_date'])
```

## Keep Only South Korea

```python
rank[ rank.country_full == "Korea Republic" ]
```

```python
rank_kor = rank[ rank.country_full == "Korea Republic" ].copy()
```

```python
rank_kor
```

## Keep Rank Date >= 2018-08-01

```python
rank_kor = rank_kor[ rank_kor.rank_date >= '2018-08-01'].reset_index(drop=True)
```

```python
rank_kor
```

## Check Ranking

```python
rank_kor[ rank_kor.rank_date == "2021-12-23" ]
```

```python
rank_kor[ rank_kor.rank_date == "2021-12-24" ]
```

```python
rank_kor[ rank_kor.rank_date == "2022-02-10" ]
```

## FIFA Ranking for Everyday

- 만약에 2021-12-24에 어떤 팀과 경기가 있었다면, 그 때 한국 피파 랭킹은?
- 우리에게 주어진 데이터는 모든 날짜별로 FIFA Ranking을 가지고 있지 않음

References:

- https://pandas.pydata.org/pandas-docs/version/2.3/reference/api/pandas.DataFrame.set_index.html
- https://pandas.pydata.org/pandas-docs/version/2.3/reference/api/pandas.DataFrame.resample.html
- https://pandas.pydata.org/pandas-docs/version/2.3/reference/api/pandas.core.groupby.DataFrameGroupBy.first.html
- https://pandas.pydata.org/pandas-docs/version/2.3/reference/api/pandas.DataFrame.ffill.html
- https://pandas.pydata.org/pandas-docs/version/2.3/reference/api/pandas.DataFrame.reset_index.html

```python
rank_kor.set_index("rank_date").resample("D").first()
```

```python
rank_kor.set_index("rank_date").resample("D").first().ffill()
```

```python
rank_kor.set_index("rank_date").resample("D").first().ffill().reset_index()
```

```python
rank_kor2 = rank_kor.set_index("rank_date").resample("D").first().ffill().reset_index().copy()
```

```python
rank_kor2.dtypes
```

```python
rank_kor2 = rank_kor2.astype(
    {
        "rank": int,
        "rank_change": int
    }
)
```

## Check Ranking

```python
rank_kor2[ rank_kor2.rank_date == "2021-12-23" ]
```

```python
rank_kor2[ rank_kor2.rank_date == "2021-12-24" ]
```

```python
rank_kor2[ rank_kor2.rank_date == "2022-02-10" ]
```
