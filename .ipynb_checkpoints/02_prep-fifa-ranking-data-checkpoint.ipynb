# Prep FIFA Ranking Data

```python
import pandas as pd
import os
```

## Load FIFA Ranking Data

```python
df_rank = pd.read_csv('../data/fifa_ranking-2022-10-06.csv')
```

## Convert to DateTime

```python
df_rank['rank_date'] = pd.to_datetime(df_rank['rank_date'])
```

## Keep Rank Date >= 2018-08-01

```python
df_rank = df_rank[ df_rank.rank_date >= '2018-08-01'].reset_index(drop=True)
```

```python
df_rank
```

## Standardize Country Names

```python
mapping = {
    'IR Iran': 'Iran',
    'Korea Republic': 'South Korea',
    'USA': 'United States'
}
```

```python
df_rank['country_full'] = df_rank['country_full'].replace(mapping)
```

## FIFA Ranking for Everyday

```python
# forward-fill rankings so that every match date has a ranking
df_rank = (
    df_rank.set_index('rank_date')
    .groupby('country_full', group_keys=False)
    .resample('D')
    .first()
    .ffill()
    .astype(
    {
        "rank": int,
        "rank_change": int
    })
)
```

```python
df_rank
```

## Save

```python
os.makedirs("intermediate", exist_ok=True)
```

```python
df_rank.to_parquet("./intermediate/rank.parquet")
```
